```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
df = pd.read_csv("bank.csv")
df
```

```python
df = df.drop('day', axis=1)
```

```python
df = df.drop('month', axis=1)
```

```python
def fillnaObjectMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['loan', 'duration', 'balance']
fillnaObjectMode(columns)
```

```python
def fillnaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['campaign','job','housing','loan','duration','balance','education']
fillnaIntMode(columns)
```

```python
def fillnaFloatMean(df, cols):
    for col in cols:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].mean())
    return df

columns = ['campaign', 'job', 'housing', 'loan', 'duration', 'balance']
df = fillnaFloatMean(df, columns)
```

```python
df.info()
```

```python
df['balance'] = df['balance'].astype('int64')
```

```python
def changeFloatToInt(cols):
    for i in cols:
        if pd.api.types.is_numeric_dtype(df[i]):
            df[i] = df[i].astype('int64')

columns = ['campaign','job','housing','loan','duration','balance','education']
changeFloatToInt(columns)
```

```python
df.info()
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))    # data_LE.fit(data)
        df[i] = data_LE.transform(df[i])
        
columns = ['loan','duration','balance']
convertObjToInt(columns)
```

```python
df.info()
```

```python
df.to_csv('processed_data.csv', index=False, header=True)
```
